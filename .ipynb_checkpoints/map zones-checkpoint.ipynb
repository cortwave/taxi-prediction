```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
 
# Get current size
fig_size = plt.rcParams["figure.figsize"]
  
# Set figure width to 12 and height to 9
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size

def draw_polygon(map_, polygon):
    for i in range(len(polygon)):
        p1 = polygon[i]
        p2 = polygon[i + 1] if i + 1 < len(polygon) else polygon[0]
        map_.drawgreatcircle(p1[1], p1[0], p2[1], p2[0], linewidth=2,color='b')

def draw_map(left_lower_corner, right_upper_corner, polygons):
    m = Basemap(llcrnrlon=left_lower_corner[1],
                llcrnrlat=left_lower_corner[0],
                urcrnrlon=right_upper_corner[1],
                urcrnrlat=right_upper_corner[0],
                rsphere=(6378137.00,6356752.3142),
                resolution='h',projection='merc')
    m.drawcoastlines()
    m.fillcontinents(color='coral',lake_color='aqua')
    m.drawmapboundary(fill_color='aqua')
    for polygon in polygons:
        draw_polygon(m, polygon)
    plt.title("Mercator Projection")
    plt.show()
    
def draw_polygons(polygons, delta = 0.005):
    min_lat, min_lon = float("inf"), float("inf")
    max_lat, max_lon = float("-inf"), float("-inf")
    for polygon in polygons:
        for point in polygon:
            min_lat = min(min_lat, point[0])
            min_lon = min(min_lon, point[1])
            max_lat = max(max_lat, point[0])
            max_lon = max(max_lon, point[1])
    ll_corner = (min_lat - delta, min_lon - delta)
    ru_corner = (max_lat + delta, max_lon + delta)
    draw_map(ll_corner, ru_corner, polygons)

    
polygons = [
    [(51.62383492610936,69.88242678801157), (51.624825387715354,69.88902551474513), (51.6293878972,69.89066302618782), (51.63296010584333,69.88570074124947), (51.63196945783619,69.8791009037959), (51.62740678761404,69.87746446213052)],
    [(53.86891860866421,27.54329505576974), (53.87178165051618,27.541728999065207), (53.87411009780087,27.546968564471857), (53.87357513811384,27.553774227832754), (53.87071193896461,27.55533952801257), (53.86838385678232,27.55009992142948)],
    [(39.82268359736405,-89.71733695758327), (39.81942979409892,-89.71475724513749), (39.817190429735355,-89.7173595865254), (39.818204695165115,-89.72254154927285), (39.82145832595108,-89.72512158452096), (39.823697863791274,-89.72251933424722)],
    [(-12.178107225017317,-54.289663860387535), (-12.176135550178646,-54.2919876497063), (-12.172151587588726,-54.29088912276847), (-12.17013927715301,-54.28746689283167), (-12.172110914195613,-54.28514312241728), (-12.17609489946478,-54.286241563033435)],
    [(-9.776282679175768,29.085522173780582), (-9.7742443824276,29.083239050326338), (-9.776177373267048,29.079814768154094), (-9.780148674734574,29.078673554132635), (-9.782187004393554,29.080956690096176), (-9.780253999669153,29.084381027574143)],
    [(-24.19351056985128,144.31616038573335), (-24.19171174765373,144.31155974133927), (-24.187546396421265,144.31139290288132), (-24.18517987717236,144.31582641460608), (-24.186978571274455,144.3204268836428), (-24.19114391271119,144.32059401630718)]
           ]

draw_polygons(polygons)
```
